```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv("C:\\Users\\HASSAN\\Downloads\\Telco-Customer-Churn.xlsx - Telco-Customer-Churn.csv")
data.head()
```

# Compare Churn count wrt Gender

```python
data1=data
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x ='Churn', data = data1, hue='gender',color='brown')
plt.title('Churn count with respect to Gender', fontsize=16)
```

# Find out how many Female Senior Citizens are there in the dataset

```python
SenrCitzn=data1[['gender','SeniorCitizen']]
SenrCitzn
Sen_Cit=SenrCitzn[(SenrCitzn['SeniorCitizen'] == 1) & (SenrCitzn['gender'] == 'Female')]
plt.figure(figsize=(6,6))
sns.countplot(Sen_Cit['gender'], color='violet',)
plt.title('Total no. of female senior citizens',fontsize=16)
plt.xlabel('Gender')
plt.ylabel('Count')
Sen_Cit['gender'].value_counts(dropna = False)
```

# Compare Tenure with Total Charges

```python
plt.figure(figsize=(8,8))
plt.scatter(data1['tenure'],data1['TotalCharges'],s=6, c='blue',marker='*')
plt.title('Plot of Tenure VS Total Charges',fontsize=18)
plt.xlabel('Tenure',fontsize=14)
plt.ylabel('Total Charges',fontsize=14)
```

# Find out which contract is preferred by senior citizens

```python
SenrCo = data1[data1["SeniorCitizen"]==1]
plt.figure(figsize=(10,10))
plt.title(" Distribution of Contract preferred by the Senior Citizens ", fontsize=14)
SenrCo.Contract.value_counts(sort=True).plot.pie(fontsize=14,startangle = 90,autopct = '%1.1f')
plt.show()
```

# Findings of Payment Method

```python
plt.figure(figsize=(12,12))
sns.countplot(data1['PaymentMethod'],hue=data1['gender'])
plt.xlabel('Payment Method',fontsize=14)
plt.ylabel('Count',fontsize=14)
plt.yticks(fontsize=10)
plt.xticks(fontsize=10)
```
